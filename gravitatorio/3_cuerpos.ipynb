# Generación de los símbolos

Se genera la totalidad de simbolos a utilizar para el desarrollo del ejercicio, correspondientes a:

$m1$, $m2$, $m3$, $G$, $M$, $L$, $t$, $x_1$, $x_2$, $x_1$, $x_3$, $y_1$, $y_2$, $y_3$, $z_1$, $z_2$, $z_3$

En este caso, es importante considerar que se generará un sistema con el Objeto 01 estático en el origen, por lo que no es necesario considerar su velocidad ni aceleración.

```python
m1,m2,m3, G,M,Ll= smp.symbols(r"m1,m2,m3, G,M,L", positive = True, real = True)

# Se utilizan estas funciones para tener la notación de Newton en las derivadas temporales.
# y se generan las funciones con dependencia implícita de "t".
from sympy.physics.mechanics import dynamicsymbols , init_vprinting

init_vprinting()

# Tiempo
t = smp.Symbol("t", positive = True) 
# Posiciones
x1,y1,z1 , x2,y2,z2 , x3,y3,z3 = dynamicsymbols(r"x_1,y_1,z_1 , x_2,y_2,z_2 , x_3,y_3,z_3", real = True) 
```

# Cantidades Físicas

Definimos los vectores de posición, velocidad y aceleración, además, la energía cinética y potencial del sistema, para finalmente conformar el Lagrangiano.

```python
# Vectores de posición
r1 = smp.Matrix([x1,y1,z1])
r2 = smp.Matrix([x2,y2,z2])
r3 = smp.Matrix([x3,y3,z3])

# Vectores de velocidad
v2 = smp.diff(r2)
v3 = smp.diff(r3)

# Vectores de aceleración
vv2 = v2.dot(v2)
vv3 = v3.dot(v3)

# Energía cinética
T = smp.Rational(1,2)*m2*vv2 + smp.Rational(1,2)*m3*vv3
# Energía potencial gravitatoria
V = -G*(m1*m2/((r1-r2).norm())+m1*m3/(r1-r3).norm() +m3*m2/(r3-r2).norm())

L = T - V
L
```

Ahora, obtenemos el Hamiltoniano del sistema, representado por

$$ \dot{q} = \frac{\partial H}{\partial p} $$

$$ \dot{p} = - \frac{\partial H}{\partial q} $$

```python
# Función encargada de calcular el Hamiltoniano a partir del Lagrangiano.
from lagrange_a_hamilton import cambio_L_H
coordenadas_generalizadas = [x2,y2,z2 , x3,y3,z3]

H, s_momentos_generalizados, momentos_generalizados = cambio_L_H(L,coordenadas_generalizadas)
H
```

```python
# Organizamos los momentos y coordenadas generalizadas en listas.
ecuaciones_dq = []
for i in range(len(coordenadas_generalizadas)):
    aux = smp.diff(H,s_momentos_generalizados[i])
    ecuaciones_dq.append(aux)

qs = [x2,y2,z2 , x3,y3,z3]

ecuaciones_dp = []
for i in range(len(s_momentos_generalizados)):
    aux = -smp.diff(H, qs[i])
    ecuaciones_dp.append(aux)
```

# Masa , longitudes y tiempos caracteristicos

Se definen valores característicos para facilitar la implementación en Python. Si trabajáramos con los datos completos, serían sumamente tardíos los cálculos. Sin embargo, estableciendo un valor de referencia para la masa, longitud, tiempo y velocidad, se simplifica el proceso de cálculo. En otras palabras, trabajaremos con proporciones.

$\bullet$ **Masa caracteristica:** Masa del Sol ($M_{sol} = 1,989 \times 10^{30} kg$)  $\Rightarrow$ $m'_{i} = \frac{m_{i} }{M}$ 

$\bullet$ **Longitud caracteristica :** Longitud promedio entre la Tierra y el Sol ($1 UA = 149. 597 .870 .700 m$) $\Rightarrow$ $\vec{r'}_{i} =\frac{\vec{r}}{L}$

$\bullet$ **Tiempo caracteristico :** $T = \left(\sqrt{\frac{GM_c}{L_c^3}}\right)^{-1}$ $\Rightarrow$ $t' = \frac{t}{T}$

```python
Tn = (smp.sqrt(G*M/Ll**3))**(-1)
Tn
```

```python
# Normalizar las coordenadas generalizadas
ecuaciones_dq_normalizadas = []

for i in range(len(ecuaciones_dq)):
	aux = ecuaciones_dq[i].subs([(m1,m1*M),(m2,m2*M),(m3,m3*M),(px2,px2*M*Ll/Tn),(py2,py2*M*Ll/Tn),
								(pz2,pz2*M*Ll/Tn),(px3,px3*M*Ll/Tn),(py3,py3*M*Ll/Tn),(pz3,pz3*M*Ll/Tn)])*Tn/Ll
	aux1 = aux
	ecuaciones_dq_normalizadas.append(aux1)

ecuaciones_dq_normalizadas
```

```python
# Normalizar los momentos generalizados (alerta de larga duración)
# Se irán imprimiendo números conforme se finalice cada resultado
ecuaciones_dp_normalizadas = []

for i in range(len(ecuaciones_dp)):
    aux = ecuaciones_dp[i].subs([(m1,m1*M),(m2,m2*M),(m3,m3*M),(x1,x1*Ll), (y1,y1*Ll),(z1,z1*Ll),(x2,x2*Ll), (y2,y2*Ll),(z2,z2*Ll),(x3,x3*Ll), (y3,y3*Ll),(z3,z3*Ll)])*Tn**2 /(M*Ll)
    aux1 = aux.simplify()
    print(i)
    ecuaciones_dp_normalizadas.append(aux1)

ecuaciones_dp_normalizadas
```

```python
# Convertimos las expresiones anteriores a expresiones numéricas

########################
dq2x_n = smp.lambdify([m2,px2],ecuaciones_dq_normalizadas[0])
dq2y_n = smp.lambdify([m2,py2],ecuaciones_dq_normalizadas[1])
dq2z_n = smp.lambdify([m2,pz2],ecuaciones_dq_normalizadas[2])

dp2x_n = smp.lambdify([m1,m2,m3, x1,y1,z1, x2,y2,z2, x3,y3,z3],ecuaciones_dp_normalizadas[0])
dp2y_n = smp.lambdify([m1,m2,m3, x1,y1,z1, x2,y2,z2, x3,y3,z3],ecuaciones_dp_normalizadas[1])
dp2z_n = smp.lambdify([m1,m2,m3, x1,y1,z1, x2,y2,z2, x3,y3,z3],ecuaciones_dp_normalizadas[2])

#########################
dq3x_n = smp.lambdify([m3,px3],ecuaciones_dq_normalizadas[3])
dq3y_n = smp.lambdify([m3,py3],ecuaciones_dq_normalizadas[4])
dq3z_n = smp.lambdify([m3,pz3],ecuaciones_dq_normalizadas[5])

dp3x_n = smp.lambdify([m1,m2,m3, x1,y1,z1, x2,y2,z2, x3,y3,z3],ecuaciones_dp_normalizadas[3])
dp3y_n = smp.lambdify([m1,m2,m3, x1,y1,z1, x2,y2,z2, x3,y3,z3],ecuaciones_dp_normalizadas[4])
dp3z_n = smp.lambdify([m1,m2,m3, x1,y1,z1, x2,y2,z2, x3,y3,z3],ecuaciones_dp_normalizadas[5])
```

```python
# Función auxiliar para la integración numérica

def dSdt(t, S, m1, m2, m3):
    # Desempaquetar las variables de estado
    [q1x, q1y, q1z, p1x, p1y, p1z, 
    q2x, q2y, q2z, p2x, p2y, p2z, 
    q3x, q3y, q3z, p3x, p3y, p3z] = S

    # Calcular las derivadas para el cuerpo 2
    dq2x = dq2x_n(m2, p2x)
    dq2y = dq2y_n(m2, p2y)
    dq2z = dq2z_n(m2, p2z)
    dp2x = dp2x_n(m1, m2, m3, q1x, q1y, q1z, q2x, q2y, q2z, q3x, q3y, q3z)
    dp2y = dp2y_n(m1, m2, m3, q1x, q1y, q1z, q2x, q2y, q2z, q3x, q3y, q3z)
    dp2z = dp2z_n(m1, m2, m3, q1x, q1y, q1z, q2x, q2y, q2z, q3x, q3y, q3z)

    # Calcular las derivadas para el cuerpo 3
    dq3x = dq3x_n(m3, p3x)
    dq3y = dq3y_n(m3, p3y)
    dq3z = dq3z_n(m3, p3z)
    dp3x = dp3x_n(m1, m2, m3, q1x, q1y, q1z, q2x, q2y, q2z, q3x, q3y, q3z)
    dp3y = dp3y_n(m1, m2, m3, q1x, q1y, q1z, q2x, q2y, q2z, q3x, q3y, q3z)
    dp3z = dp3z_n(m1, m2, m3, q1x, q1y, q1z, q2x, q2y, q2z, q3x, q3y, q3z)

    # Retornar la lista de derivadas
    return [q1x, q1y, q1z, p1x, p1y, p1z, 
            dq2x, dq2y, dq2z, dp2x, dp2y, dp2z, 
            dq3x, dq3y, dq3z, dp3x, dp3y, dp3z]
```

# Sistema de estudio

```python
# Masas, se define la masa del Sol ('masa_Sol') como masa característica
masa_Sol = 1.98930e30 #kg
masa_Tierra = 5.9722e24 #kg
masa_Luna = 7.3458e22 #kg


# Longitud característica
UA = 149597870700 #m

T_n = Tn.subs([(Ll,UA),(M,masa_Sol),(G,6.672 *10**(-11))]).evalf()
```

```python
# Durante esta sección, se trabajará con velocidades y distancias promedio.

t0 = 0
tf = 3600*24*365/T_n
n=round((tf-t0)/(3600*24/T_n))

t_n = np.linspace(t0,int(tf),n)

# Argumentos

# Definimos las masas relativas
m1_n = 1 #Sol
m2_n = masa_Luna/masa_Sol #Luna
m3_n = masa_Tierra/masa_Sol #Tierra

argumentos = [m1_n,m2_n,m3_n]

# Condiciones iniciales

####################### Sol

# Definimos las longitudes relativas

# El Sol lo definimos estático en el Origen, r1=(0,0,0)

q1x0 = 0
q1y0 = 0
q1z0 = 0

p1x0 = 0
p1y0 = 0
p1z0 = 0

####################### Tierra

# Las expresiones de las órbitas estables son derivadas de las Leyes de Kepler.
# Se necesita la masa del objeto al que orbitan y la distancia (promedio) entre ambos cuerpos.

G = 6.67430e-11 #m³/kg s
# Distancia de la Tierra a la Luna
D_TL = 3.84e8  # m
# Velocidad de la Tierra alrededor del Sol para asegurar una órbita estable.
v_T = np.sqrt(G * masa_Sol / UA)
# Velocidad de la Luna alrededor de la Tierra para asegurar una órbita estable.
v_L = np.sqrt(G * masa_Tierra / D_TL)

# La Tierra se encuentra a una Unidad Astronómica (UA), iniciamos en X=UA, Y=0, Z=0
q3x0 = UA/UA
q3y0 = 0
q3z0 = 0

# Definimos la velocidad inicial como perpendicular al vector posición inicial.
p3x0 = 0
p3y0 = m3_n*v_T/v_T
p3z0 = 0

####################### Luna

# La distancia promedio de la Tierra a la Luna corresponde a 384.399.860m
# Iniciamos el sistema alineando la Luna 'detrás' de la Tierra.
# Considerando la velocidad promedio de rotación de la Tierra alrededor del Sol,
# la velocidad de la Luna respecto al Sol estará dada, inicialmente, sumándole la
# velocidad promedio de rotación de la Luna alrededor de la Tierra.

# Es necesario considerar que la órbita de la Luna está inclinada 5° respecto al plano de la órbita de la Tierra.

# Sumar la distancia del Sol a la Tierra (UA) con la distancia de la Tierra a la Luna.
q2x0 = 1+(D_TL/UA)*np.cos(np.radians(5))
q2y0 = 0
q2z0 = (D_TL/UA)*np.sin(np.radians(5))

# Para calcular el momento lineal, se suma la velocidad dada por la órbita de la Luna, inicialmente
# perpendicular al vector posición inicial, con la velocidad dada por la órbita de la Tierra.
# Definimos la velocidad de la Tierra como velocidad característica, para simplificar los cálculos.
p2x0 = 0
p2y0 = m2_n*(v_T+v_L*np.cos(np.radians(5)))/v_T
p2z0 = m2_n*v_L*np.sin(np.radians(5))/v_T



condiciones_iniciales = [q1x0,q1y0,q1z0, p1x0,p1y0,p1z0, q2x0,q2y0,q2z0, p2x0,p2y0,p2z0, q3x0,q3y0,q3z0, p3x0,p3y0,p3z0]

argumentos
```

```python
from scipy.integrate  import solve_ivp


sol = solve_ivp(dSdt, t_span = (t0,tf), t_eval = t_n, y0 = condiciones_iniciales, args = argumentos, method='LSODA', rtol=1e-10, atol=1e-12)
pos = sol.y
```

```python
# Generar una imagen del estado final del sistema, en el intervalo de tiempo indicado previamente (ver 't_n')

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Graficar trayectorias
planet1_plt, = ax.plot(pos[0], pos[1], pos[2], 'orange', label='Sol', linewidth=1)
planet2_plt, = ax.plot(pos[6], pos[7], pos[8], 'black', label='Luna', linewidth=1)
planet3_plt, = ax.plot(pos[12], pos[13], pos[14], 'blue', label='Tierra', linewidth=1)

# Graficar puntos finales
planet1_dot, = ax.plot([pos[0][-1]], [pos[1][-1]], [pos[2][-1]], 'o', color='orange', markersize=7)
planet2_dot, = ax.plot([pos[6][-1]], [pos[7][-1]], [pos[8][-1]], 'o', color='black', markersize=7)
planet3_dot, = ax.plot([pos[12][-1]], [pos[13][-1]], [pos[14][-1]], 'o', color='blue', markersize=7)

# Configurar títulos y etiquetas
ax.set_title("The 3-Body Problem")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.grid()
#plt.legend()

# Asegurar que el origen esté centrado y la escala sea fija
max_range = max(
    max(abs(pos[0])), max(abs(pos[1])), max(abs(pos[2])),
    max(abs(pos[6])), max(abs(pos[7])), max(abs(pos[8])),
    max(abs(pos[12])), max(abs(pos[13])), max(abs(pos[14]))
)

ax.set_xlim([-max_range, max_range])
ax.set_ylim([-max_range, max_range])
ax.set_zlim([-max_range, max_range])

# Hacer que la escala sea isótropa
ax.set_box_aspect([1, 1, 1])  # Relación de aspecto isótropa

# Guardar la figura
plt.savefig("3_cuerpos.png")
```

```python
# Generar la animación correspondiente a la evolución del sistema
# El resultado se guarda en la carpeta en que se encuentra este archivo, con el nombre "3_cuerpos_animacion.mp4"

from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from datetime import datetime, timedelta

# Inicializar el tiempo
initTime = datetime.now()

# Crear la figura y el eje 3D
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Configurar las etiquetas
ax.set_title("The 3-Body Problem")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

# Inicializar las trayectorias y puntos
planet1_plt, = ax.plot([], [], [], 'orange', label='Sol', linewidth=1)
planet2_plt, = ax.plot([], [], [], 'black', label='Luna', linewidth=1)
planet3_plt, = ax.plot([], [], [], 'blue', label='Tierra', linewidth=1)

planet1_dot, = ax.plot([], [], [], 'o', color='orange', markersize=7)
planet2_dot, = ax.plot([], [], [], 'o', color='black', markersize=7)
planet3_dot, = ax.plot([], [], [], 'o', color='blue', markersize=7)

# Agregar una caja de texto para mostrar el tiempo
time_text = ax.text2D(0.05, 0.95, '', transform=ax.transAxes, fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

# Número de puntos recientes a considerar para los límites
window = 100
smooth_factor = 0.2  # Factor de suavizado para ajustar la expansión de los límites

# Variables para almacenar los límites actuales
current_xlim = [-1, 1]
current_ylim = [-1, 1]
current_zlim = [-1, 1]
expanded = False  # Estado que indica si los límites ya se han expandido

# Función para ajustar dinámicamente los límites solo cuando sea necesario
def adjust_limits(i):
    global current_xlim, current_ylim, current_zlim, expanded
    
    # Asegurar que el índice de la ventana no sea negativo
    start_idx = max(0, i - window)

    # Obtener las posiciones de los objetos
    x_all = np.concatenate([pos[0][start_idx:i], pos[6][start_idx:i], pos[12][start_idx:i]])
    y_all = np.concatenate([pos[1][start_idx:i], pos[7][start_idx:i], pos[13][start_idx:i]])
    z_all = np.concatenate([pos[2][start_idx:i], pos[8][start_idx:i], pos[14][start_idx:i]])

    # Si los arreglos están vacíos, retornar
    if x_all.size == 0 or y_all.size == 0 or z_all.size == 0:
        return

    # Calcular los nuevos límites con un margen adicional
    x_min, x_max = x_all.min(), x_all.max()
    y_min, y_max = y_all.min(), y_all.max()
    z_min, z_max = z_all.min(), z_all.max()

    # Calcular el rango máximo entre los tres ejes para hacer la escala simétrica
    max_range = max(x_max - x_min, y_max - y_min, z_max - z_min)
    margin = 0.1 * max_range  # Margen para asegurarse de que hay espacio alrededor de los puntos

    # El Sol siempre estará en el centro de la imagen, por lo que los límites se ajustan respecto a él.
    center_x = 0  # El Sol está en el centro de la imagen
    center_y = 0
    center_z = 0

    # Calcular los nuevos límites en torno al Sol y suavizar la expansión
    new_xlim = [center_x - max_range / 2 - margin, center_x + max_range / 2 + margin]
    new_ylim = [center_y - max_range / 2 - margin, center_y + max_range / 2 + margin]
    new_zlim = [center_z - max_range / 2 - margin, center_z + max_range / 2 + margin]

    # Si no se ha expandido antes, o si es necesario expandir más, ajustamos los límites
    if not expanded or (
        new_xlim[0] < current_xlim[0] or new_xlim[1] > current_xlim[1] or
        new_ylim[0] < current_ylim[0] or new_ylim[1] > current_ylim[1] or
        new_zlim[0] < current_zlim[0] or new_zlim[1] > current_zlim[1]
    ):
        # Suavizar la transición entre los límites actuales y los nuevos límites
        current_xlim = [current_xlim[0] + smooth_factor * (new_xlim[0] - current_xlim[0]),
                        current_xlim[1] + smooth_factor * (new_xlim[1] - current_xlim[1])]
        
        current_ylim = [current_ylim[0] + smooth_factor * (new_ylim[0] - current_ylim[0]),
                        current_ylim[1] + smooth_factor * (new_ylim[1] - current_ylim[1])]
        
        current_zlim = [current_zlim[0] + smooth_factor * (new_zlim[0] - current_zlim[0]),
                        current_zlim[1] + smooth_factor * (new_zlim[1] - current_zlim[1])]
        
        # Actualizamos el estado de expansión
        expanded = True

    # Establecer los nuevos límites suavizados
    ax.set_xlim(current_xlim)
    ax.set_ylim(current_ylim)
    ax.set_zlim(current_zlim)

# Función de inicialización
def init():
    planet1_plt.set_data([], [])
    planet1_plt.set_3d_properties([])
    
    planet2_plt.set_data([], [])
    planet2_plt.set_3d_properties([])
    
    planet3_plt.set_data([], [])
    planet3_plt.set_3d_properties([])
    
    planet1_dot.set_data([], [])
    planet1_dot.set_3d_properties([])
    
    planet2_dot.set_data([], [])
    planet2_dot.set_3d_properties([])
    
    planet3_dot.set_data([], [])
    planet3_dot.set_3d_properties([])
    
    time_text.set_text('')  # Inicializar la caja de texto vacía
    
    # Establecer límites iniciales basados en la distancia inicial de la Luna (D_TL)
    distance_Luna = 3.84e8  # Distancia inicial entre la Tierra y la Luna
    margin = 0.1 * distance_Luna
    ax.set_xlim([-distance_Luna - margin, distance_Luna + margin])
    ax.set_ylim([-distance_Luna - margin, distance_Luna + margin])
    ax.set_zlim([-distance_Luna - margin, distance_Luna + margin])
    
    return planet1_plt, planet2_plt, planet3_plt, planet1_dot, planet2_dot, planet3_dot, time_text

# Función de animación
def animate(i):
    global initTime
    global tf
    global n
    global T_n
    
    # Ajustar dinámicamente los límites
    adjust_limits(i)
    
    # Actualizar trayectorias
    planet1_plt.set_data(pos[0][:i], pos[1][:i])
    planet1_plt.set_3d_properties(pos[2][:i])

    planet2_plt.set_data(pos[6][:i], pos[7][:i])
    planet2_plt.set_3d_properties(pos[8][:i])

    planet3_plt.set_data(pos[12][:i], pos[13][:i])
    planet3_plt.set_3d_properties(pos[14][:i])

    # Actualizar los puntos de posición actual
    planet1_dot.set_data([pos[0][i]], [pos[1][i]])
    planet1_dot.set_3d_properties([pos[2][i]])

    planet2_dot.set_data([pos[6][i]], [pos[7][i]])
    planet2_dot.set_3d_properties([pos[8][i]])

    planet3_dot.set_data([pos[12][i]], [pos[13][i]])
    planet3_dot.set_3d_properties([pos[14][i]])

    # Actualizar el tiempo correspondiente usando sol.t
    newTime = initTime + timedelta(seconds=int((sol.t[i] - sol.t[0])*T_n))  # Tiempo inicial ajustado

    # Actualizar la caja de texto con el tiempo correspondiente
    time_text.set_text(newTime.strftime('%d %b %Y'))

    return planet1_plt, planet2_plt, planet3_plt, planet1_dot, planet2_dot, planet3_dot, time_text


# Crear la animación
frames_total = len(pos[0])  # Total de frames en los datos
desired_duration = 60  # Duración deseada en segundos
fps = round(frames_total / desired_duration)  # Calcular FPS necesario

ani = FuncAnimation(fig, animate, frames=frames_total, init_func=init, blit=True)

# Guardar como video
ani.save("3_cuerpos_animacion.mp4", writer="ffmpeg", fps=fps*3)
```
